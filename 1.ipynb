````python
import numpy as np 
import matplotlib.pyplot as plt
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
````

Build a linear regression model to fit the above generated data and plot thefunction. Also calculate MSE and R2 score of the fitting.

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
````

````python
x.shape
````

````python
y.shape
````

````python
train_X, test_X, train_y, test_y = train_test_split(x.reshape(-1, 1), y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(train_X, train_y)
y_pred = model.predict(test_X)
````

````python
mean_squared_error(test_y, y_pred), r2_score(test_y, y_pred)
````

````python
y_pred = model.predict(x.reshape(-1, 1))
````

````python
plt.scatter(x, y);
plt.plot(x, y_pred, color='red');
````

Build a linear-SVM regression on the above generated data and plot thefunction. Also calculate MSE and R2 score of the fitting.

````python
from sklearn.svm import SVR
svr_model = SVR(kernel='linear', C=100, epsilon=1.5)
svr_model.fit(train_X, train_y)
y_svr_pred = svr_model.predict(test_X)
````

````python
mean_squared_error(test_y, y_svr_pred), r2_score(test_y, y_svr_pred)
````
